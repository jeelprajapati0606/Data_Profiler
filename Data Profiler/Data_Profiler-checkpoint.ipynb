## Q4. Tensors

### (a). What are Tensors?
#### Answer:

#### A Tensor is a multi-dimensional array used to store numerical data.

#### Types of Tensors:

| Type | Dimension | Example |
|---|---|---|
| Scalar | 0D | `5` |
| Vector | 1D | `[1, 2, 3]` |
| Matrix | 2D | `[[1, 2], [3, 4]]` |
| 3D Tensor | 3D | Multiple matrices |

### (B) Provide an in-depth explanation of tensors with simple examples using NumPy.

```python
import numpy as np

# Scalar
scalar = np.array(5)

# Vector
vector = np.array([1, 2, 3])

# Matrix
matrix = np.array([
    [1, 2],
    [3, 4]
])

# 3D Tensor
tensor_3d = np.array([
    [
        [1, 2],
        [3, 4]
    ],
    [
        [5, 6],
        [7, 8]
    ]
])

print("Scalar Dimension:", scalar.ndim)
print("Vector Dimension:", vector.ndim)
print("Matrix Dimension:", matrix.ndim)
print("3D Tensor Dimension:", tensor_3d.ndim)
```

## Part B: Data Acquisition

### Q5. Import datasets from different sources

#### (a). Import CSV using Pandas

```python
import pandas as pd

data = pd.read_csv("customer_basic_data.csv")
data.head()
```

### (b). Parse JSON Data

```python
import json
import pandas as pd

with open("purchase_data.json","r") as file :
    json_data = json.load(file)

data_json = pd.DataFrame(json_data)

data_json.head()
```

## (c). Load Data from SQL

```python
import mysql.connector
import pandas as pd

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="customers"
)

sql_data = pd.read_sql(
    "SELECT * FROM customers",
    conn
)

print(sql_data.head())

conn.close()
```

## (d). Load Data from API


```python
import json
import pandas as pd

with open("api_customer_data.json", "r") as file :
    api_data = json.load(file)

data_api =  pd.DataFrame(api_data)
data_api.head()
```

# Merge All Data Sources :

```python
# Merge all data sources

df = data.merge(
    data_json,
    on="Customer_ID",
    how="left"
)

df = data.merge(
    sql_data,
    on="Customer_ID",
    how="left"
)

df = data.merge(
    data_api,
    on="Customer_ID",
    how="left"
)

```

## Q7. Apply Data Cleaning

### (a). Handle Missing Values

```python
# For Numerical Colunms :

df["Income"] = df["Income"].fillna(
    df["Income"].median()
)

# For categorical columns:

df["Gender"] = df["Gender"].fillna(
    df["Gender"].mode()[0]
)
```

### (b). Correct Data Types

```python
df["Age"] = pd.to_numeric(
    df["Age"],
    errors="coerce"
)

df["Income"] = pd.to_numeric(
    df["Income"],
    errors="coerce"
)
```

### (c). Remove Duplicate Rows

```python
df = df.drop_duplicates()
```

# Part D: Exploratory Data Analysis (EDA)

### Q8. Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 5))

# Age
plt.subplot(1, 3, 1)
sns.histplot(df["Age"], kde=True)
plt.title("Distribution of Customer Age")
plt.xlabel("Age")
plt.ylabel("Number of Customers")

# Income
plt.subplot(1, 3, 2)
sns.histplot(df["Income"], kde=True)
plt.title("Distribution of Customer Income")
plt.xlabel("Income")
plt.ylabel("Number of Customers")

# Purchases
plt.subplot(1, 3, 3)

sns.histplot(data_json["Purchases"], kde=True)
plt.title("Distribution of Purchases")
plt.xlabel("Number of Purchases")
plt.ylabel("Number of Customers")

plt.tight_layout()
plt.show()
```

## Q9. Bivariate Analysis

### (a). Relationship Between Gender and Purchases

```python
sns.boxplot(
    data=df,
    x="Gender",
    y=data_json["Purchases"]
)

plt.title("Gender vs Purchases")
plt.show()
```

### (b). Relationship Between Income and Churn

```python
sns.boxplot(
    data=df,
    x=sql_data["Churn"],
    y="Income"
)

plt.title("Income vs Churn")
plt.show()
```

## Q10. Multivariate Analysis

### (a). Correlation Heatmap

```python
numeric_df = df.select_dtypes(
    include="number"
)

plt.figure(figsize=(10, 7))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

### (b). Pair Plot

```python
pairplot_df = pd.concat(
    [
        df[["Age", "Income"]].reset_index(drop=True),
        data_json[["Purchases", "Total_Spending"]].reset_index(drop=True),
        sql_data[["Last_Purchase_Days", "Churn"]].reset_index(drop=True)
    ],
    axis=1
)

pairplot_df.head()
```

```python
sns.pairplot(
    pairplot_df,
    hue="Churn"
)

plt.show()
```
